#### Outlier Analysis

```python
sns.boxplot(y='degree_p',data=df)
```

```python
percentile25 = df['degree_p'].quantile(0.25)
percentile75 = df['degree_p'].quantile(0.75)
```

```python
IQR = percentile75-percentile25
```

```python
upper_limit = percentile75+1.5*IQR
lower_limit = percentile25-1.5*IQR
```

```python
df = df[df['degree_p']<upper_limit]
df = df[df['degree_p']>lower_limit]
```

```python
sns.boxplot(y='degree_p',data=df)
```

```python
df.columns
```

```python
col = ['ssc_p','hsc_p','degree_p','etest_p','mba_p','salary']    
```

```python
for c in col:
    plt.figure()
    sns.boxplot(y=c,data=df)
```

```python
#iterating outlier analysis for all countinous columns
for c in col:
    percentile25 = df[c].quantile(0.25)
    percentile75 = df[c].quantile(0.75)
    IQR = percentile75-percentile25
    upper_limit = percentile75+1.5*IQR
    lower_limit = percentile25-1.5*IQR
    df = df[df[c]<upper_limit]
    df = df[df[c]>lower_limit]
```

```python
for c in col:
    plt.figure()
    sns.boxplot(y=c,data=df)
```

```python
new_df.head()
```

```python
#Evaluation using Inertia
kmeans.inertia_
```

```python
new_df1 = df
new_df1.shape
```

```python
#Finding Optimal number of clusters using Elbow Method
inertia = []
for i in range(1,204):
    kmeans = KMeans(i)
    kmeans.fit(new_df1)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
clusters = range(1,204)
```

```python
plt.figure(figsize=(20,10))
plt.plot(clusters,inertia,marker='o')
plt.xlabel("Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Curve")
```

#### Optimal Number of clusters would be 2 or 3

### Principal Component Analysis

```python
df
```

```python
xx = df.reset_index().drop(columns = ['status','salary'],axis=1).values
```

```python
xx.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Standardization
xx_trans = StandardScaler().fit_transform(xx)
```

```python
#Importing Principal Component Analysis
from sklearn.decomposition import PCA
```

```python
#calling for 2 Principal Componennts
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(xx_trans)
```

```python
principal_components
```

```python
#converting 2darray into dataframe
pc_df = pd.DataFrame(data = principal_components,columns=['PC1','PC2'])
```

```python
#concatenate pc_df to our dataframe
final_pc_df = pd.concat([pc_df,df.reset_index()[['status','salary']]],axis =1)
```

#### Now the final Principal Component dataframe is ready for Linear or Logistic regression
